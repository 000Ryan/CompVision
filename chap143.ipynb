```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('ryan.jpg')
resC = cv2.resize(img,None,fx=2,fy=2,interpolation = cv2.INTER_CUBIC)
resN = cv2.resize(img,None,fx=2,fy=2,interpolation = cv2.INTER_NEAREST)
```

```python
cv2.imshow('image',img)
cv2.imshow('resimgc',resC)
cv2.imshow('resimgN',resN)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#or
height, width = img.shape[:2]
res = cv2.resize(img,(2*width,2*height), interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image',img)
cv2.imshow('resimg',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Translation

```python
img = cv2.imread('messi5.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('image',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('messi5.jpg',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('image',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('ryan.jpg')
rows,cols,ch = img.shape



pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('ryan.jpg')
rows,cols,ch = img.shape
cx,cy=250,150
theta = 300
M = cv2.getRotationMatrix2D((cy,cx),theta,1)
dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.circle(dst,(cy,cx),10,(0,0,255),-1)
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[235,457],[565,492],[565,234]])
pts2 = np.float32([[235,457],[568,457],[568,269]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('image',dst)
cv2.imshow('image1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[235,457],[238,269],[569,231],[564,493]])
pts2 = np.float32([[235,457],[235,270],[570,270],[570,457]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
