```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)

plt.subplot(121),plt.imshow(erosion),plt.title('Erosion')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 2)

plt.subplot(121),plt.imshow(dilation),plt.title('Dilation')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121),plt.imshow(opening),plt.title('Opening')
plt.subplot(122),plt.imshow(img),plt.title('original')
plt.xticks([]), plt.yticks([])
plt.show()

cv2.imshow('dst',opening)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121),plt.imshow(closing),plt.title('Closing')
plt.subplot(122),plt.imshow(img),plt.title('original')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imshow('dst',opening)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
